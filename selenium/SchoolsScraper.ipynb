```python
schools = pd.read_csv('eani.csv')
schools
```

```python
driver = webdriver.Chrome()
for school in schools["entry href"]:
    driver.get(school)
    driver.implicitly_wait(3)
    name = driver.find_element(By.XPATH, '//*[@id="block-eani-content"]/div/div[1]/div/div/h1').text
    about_school = driver.find_elements(By.XPATH, '//*[@id="block-eani-content"]/div/div[2]/div[1]/div[2]/div')
#     school_name.append(name)
#     dne_number.append(about_school[0].text.split(":")[1])
#     school_type.append(about_school[1].text.split(":")[1])
#     age_range.append(about_school[4].text.split(":")[1])
#     parl_const.append(about_school[7].text.split(":")[1]) 
    school_details.append(f'{name}, {[detail.text for detail in about_school]}')
    time.sleep(1)
```
